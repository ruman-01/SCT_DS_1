```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the main dataset (skipping the first 4 rows)
df_main = pd.read_csv('API.csv', skiprows=4)

# Load the metadata to filter out regional aggregates
df_meta = pd.read_csv('Metadata_Country_API.csv')

# Merge the datasets
df_merged = pd.merge(df_main, df_meta, on='Country Code', how='inner')

# Keep only actual countries (Regions have 'NaN' in the Region column)
df_countries = df_merged.dropna(subset=['Region'])
```

```python
# 1. Sorting and getting the top 10 countries for the year '2022'
top_10 = df_countries.nlargest(10, '2022')

# 2. Setting up the figure size
plt.figure(figsize=(10, 6))

# 3. Creating a horizontal barplot
sns.barplot(x='2022', y='Country Name', data=top_10,hue='Country Name',palette='viridis')

# 4. Adding titles and labels to make it professional
plt.title('Top 10 Most Populous Countries in 2022')
plt.xlabel('Total Population (in billions)')
plt.ylabel('Country')
plt.tight_layout()

# 5. Showing the plot
plt.show()
```
